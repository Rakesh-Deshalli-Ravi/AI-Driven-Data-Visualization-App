```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
plt.style.use('seaborn-v0_8')
```

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
```

```python
plt.scatter(x,y)
plt.tight_layout()
plt.show()
```

```python
plt.scatter(x,y, s = 100, c = 'green',alpha = 0.75,
            edgecolor = 'black',linewidth = 1)
plt.tight_layout()
plt.show()
```

```python
colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]
plt.scatter(x,y, s = 100, c = colors,alpha = 0.75,
            edgecolor = 'black',linewidth = 1)
plt.tight_layout()
plt.show()
```

```python
colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]
plt.scatter(x,y, s = 100, c = colors, cmap = 'Greens',alpha = 0.75,
            edgecolor = 'black',linewidth = 1)
plt.tight_layout()
plt.show()
```

```python
colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]
plt.scatter(x,y, s = 100, c = colors, cmap = 'Greens',alpha = 0.75,
            edgecolor = 'black',linewidth = 1)
cbar = plt.colorbar()
cbar.set_label('satisfaction level')
plt.tight_layout()
plt.show()
```

```python
data = pd.read_csv('data_3.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']
```

```python
plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')
plt.scatter(view_count,likes,alpha = 0.75,
            edgecolor = 'black',linewidth = 1)
```

```python
plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')
plt.scatter(view_count,likes,alpha = 0.75,
            edgecolor = 'black',linewidth = 1)
plt.xscale('log')
plt.yscale('log')
```

```python
plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')
plt.scatter(view_count,likes,alpha = 0.75, c = ratio, cmap = 'summer',
            edgecolor = 'black',linewidth = 1)
cbar = plt.colorbar()
cbar.set_label('like/dislike ration')
plt.xscale('log')
plt.yscale('log')
plt.tight_layout()
plt.show()
```
